```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Getting the dataset
dataset=pd.read_csv('C:\\Neville\\Python\\Session\\Module 8\\breast-cancer-data.csv')
```

```python
#Checking the top 5 rows
dataset.head()
```

```python
#Check if there are any null values
dataset.isnull().sum()
```

```python
#Check the datatypes of all fields
dataset.info()
```

```python
#Get all the independent variables
x=dataset.iloc[:,2:].values
```

```python
#Get the dependent variable
y=dataset.iloc[:,1:2].values
```

```python
#Convert all the independent variables to the same scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x[:]=sc.fit_transform(x[:])
```

```python
#Having a look at the independent variables post feature scaling
pd.DataFrame(x)
```

```python
#Convert the target variable to a numerical format using Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y[:])
```

```python
#Have a look at the target variable post lable encoding
y
```

```python
#Split the data for training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#Use LogisticRegression to fit the model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Perform prediction
y_pred=model.predict(x_test)
```

```python
#Check the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Perform PCA to reduce the number of dimensions
from sklearn.decomposition import PCA
pca=PCA(n_components=None)
x_train_n=pca.fit_transform(x_train)
x_test_n=pca.fit_transform(x_test)
```

```python
#Find the variance factor of each variable
pca.explained_variance_ratio_
```

```python
#Perform PCA using 2 components
pca=PCA(n_components=2)
x_train_2=pca.fit_transform(x_train)
x_test_2=pca.fit_transform(x_test)
```

```python
#Fit the model using the 2 new pca created above
model.fit(x_train_2,y_train)
```

```python
#Predict using the reduced components
y_pred_2=model.predict(x_test_2)
```

```python
#Check the accuracy again
accuracy_score(y_test,y_pred_2)
```
