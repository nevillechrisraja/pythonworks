```python
#Convert the list of independent variables to an array
x=dataset.iloc[:,1:-1].values
```

```python
#Convert the dependent variable to an array
y=dataset.iloc[:,-1]
```

```python
#Split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#Fit the model using Linear Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
#Check the accuracy
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#Check the accuracy using Adjusted R2 
import statsmodels.api as sm
regressor_ols=sm.OLS(exog=x,endog=y).fit()
regressor_ols.summary()
```

```python
#Use only the independent variables with p value < 0.05
x=x[:,[0,1,2,3,6,7,10,12,13,14,19,24,25,26,27,28,43,45]]
```

```python
#Again check the adjusted R2
regressor_ols2=sm.OLS(exog=x,endog=y).fit()
regressor_ols2.summary()
```

```python
#Split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train2,x_test2,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#Convert all the independent variables to the same scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train3=sc.fit_transform(x_train2[:,:])
```

```python
#Again check the adjusted R2 after StandardScaler
regressor_ols3=sm.OLS(exog=x_train3,endog=y_train).fit()
regressor_ols3.summary()
```

```python
#Fit & Predict the model again using Linear Regression with x_train2 and x_test2
model.fit(x_train2,y_train)
y_pred=model.predict(x_test2)
```

```python
#Fit the model with Random Forest
from sklearn.ensemble import RandomForestRegressor
model_rf=RandomForestRegressor(n_estimators=50)
model_rf.fit(x_train2,y_train)
y_pred_rf=model_rf.predict(x_test2)
```

```python
#Fit the model with Decision Tree
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train2,y_train)
y_pred_dt=model.predict(x_test2)
```

```python
#Root Mean Square
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_test, y_pred))
rms_rf = sqrt(mean_squared_error(y_test, y_pred_rf))
rms_dt = sqrt(mean_squared_error(y_test, y_pred_dt))
print("Linear Regression:")
print(rms)
print("Random Forest:")
print(rms_rf)
print("Decision Tree:")
print(rms_dt)
```
