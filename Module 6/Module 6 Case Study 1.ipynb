```python
#Question 2a
dataset['total_benefitted']=dataset['No. of Inmates benefitted by Elementary Education']+dataset['No. of Inmates benefitted by Adult Education']+dataset['No. of Inmates benefitted by Higher Education']+dataset['No. of Inmates benefitted by Computer Course']
```

```python
#Question 2b
total_ee=dataset['No. of Inmates benefitted by Elementary Education'].sum()
total_ae=dataset['No. of Inmates benefitted by Adult Education'].sum()
total_he=dataset['No. of Inmates benefitted by Higher Education'].sum()
total_cc=dataset['No. of Inmates benefitted by Computer Course'].sum()
```

```python
dataset=dataset.append({'STATE/UT':'total','No. of Inmates benefitted by Elementary Education':total_ee,
         'No. of Inmates benefitted by Adult Education':total_ae,
          'No. of Inmates benefitted by Higher Education':total_he,
          'No. of Inmates benefitted by Computer Course':total_cc
         },ignore_index=True)
```

```python
#Question 3a
subsetdata=dataset.iloc[:35,:6]
subsetdata=dataset.sort_values(['total_benefitted'],ascending=0).head(5)
sns.barplot(data=subsetdata,x='STATE/UT',y='total_benefitted')
plt.show()
```
