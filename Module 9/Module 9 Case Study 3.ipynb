```python
#Question 2: Use LabelEncoder to encode the target variable in to numerical form and split the data such that 20% of the data is set aside for testing.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Private']=le.fit_transform(dataset['Private'])
```

```python
y_pred=model.predict(x_test)
```
